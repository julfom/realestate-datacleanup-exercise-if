#### Exercise 09. Are the average prices of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about them.

```python
# TODO
import pandas as pd
valdemorillo = ds[ds['address'] == "Valdemorillo"]
valde_prices = valdemorillo['price'].dropna()
valde_average_price = valde_prices.mean()
print(f"The average price for entries with address 'Valdemorillo' is: {valde_average_price}")

galapagar = ds[ds['address'] == "Galapagar"]
gala_prices = galapagar['price'].dropna()
gala_average_price = gala_prices.mean()
print(f"The average price for entries with address 'Galapagar' is: {gala_average_price}")

if gala_average_price > valde_average_price:
    print('Average price of Galapagar is higher'),
else: print('Average price of Valdemorillo is higher')
```
